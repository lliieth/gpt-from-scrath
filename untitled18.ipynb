```python
import json

terms = {
    "malware": "Malware is malicious software designed to damage or exploit systems.",
    "phishing": "Phishing is a social engineering attack used to steal user credentials.",
    "ransomware": "Ransomware is malware that encrypts files and demands payment.",
    "firewall": "A firewall monitors and controls incoming and outgoing network traffic.",
    "encryption": "Encryption protects data by converting it into unreadable form.",
    "vulnerability": "A vulnerability is a weakness in a system that can be exploited.",
    "exploit": "An exploit is code or a method used to take advantage of a vulnerability.",
    "authentication": "Authentication verifies the identity of a user or device.",
    "botnet": "A botnet is a group of infected devices controlled by an attacker.",
    "rootkit": "A rootkit is malicious software designed to hide itself inside a system."
}

arabic_texts = [
    "الأمن السيبراني يحمي الأنظمة والشبكات من الهجمات.",
    "التشفير يحمي البيانات الحساسة من الوصول غير المصرح به.",
    "التصيد الاحتيالي يستخدم لخداع المستخدم وسرقة بياناته.",
    "الجدار الناري يراقب حركة المرور داخل الشبكة.",
    "البرمجيات الخبيثة قد تسبب ضررًا كبيرًا للأنظمة."
]

english_texts = [
    "Cybersecurity includes technologies and processes used to protect systems and networks.",
    "Phishing attacks rely on deception to steal sensitive information.",
    "Encryption is important because it protects confidential data.",
    "A firewall helps control network traffic and blocks suspicious connections.",
    "Malware can damage files, steal data, or disrupt services."
]

sft_data = []

# Q&A
for term, definition in terms.items():
    sft_data.append({
        "instruction": f"What is {term}?",
        "input": "",
        "output": definition
    })

# Explanation
for term, definition in terms.items():
    sft_data.append({
        "instruction": f"Explain {term} in simple words.",
        "input": "",
        "output": definition
    })

# Translation EN -> AR
for term, definition in terms.items():
    sft_data.append({
        "instruction": "Translate to Arabic",
        "input": definition,
        "output": "ترجمة تعريف المصطلح إلى العربية."
    })

# Translation AR -> EN
for ar in arabic_texts:
    sft_data.append({
        "instruction": "Translate to English",
        "input": ar,
        "output": "Translation of the Arabic cybersecurity sentence into English."
    })

# Summarization
for text in english_texts:
    sft_data.append({
        "instruction": "Summarize the following text",
        "input": text,
        "output": "A short summary of the cybersecurity concept."
    })

# Arabic Q&A
for ar in arabic_texts:
    sft_data.append({
        "instruction": "اشرح الجملة التالية",
        "input": ar,
        "output": "هذه الجملة تشرح مفهومًا أساسيًا في الأمن السيبراني."
    })

# كرر حتى نوصل 200
base = list(sft_data)
while len(sft_data) < 200:
    for item in base:
        if len(sft_data) >= 200:
            break
        new_item = item.copy()
        new_item["instruction"] = item["instruction"] + f" #{len(sft_data)+1}"
        sft_data.append(new_item)

with open("sft_data.json", "w", encoding="utf-8") as f:
    json.dump(sft_data[:200], f, ensure_ascii=False, indent=2)

print("SFT file created successfully.")
print("Total instructions:", len(sft_data[:200]))
```

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

with open("cyber_dataset.txt", "r", encoding="utf-8") as f:
    text = f.read()

chars = sorted(list(set(text)))
vocab_size = len(chars)

stoi = {ch:i for i,ch in enumerate(chars)}
itos = {i:ch for i,ch in enumerate(chars)}

encode = lambda s: [stoi[c] for c in s]
decode = lambda l: ''.join([itos[i] for i in l])

data = torch.tensor(encode(text), dtype=torch.long)

class TinyGPT(nn.Module):
    def __init__(self, vocab):
        super().__init__()
        self.embed = nn.Embedding(vocab, 64)
        self.linear = nn.Linear(64, vocab)

    def forward(self, x):
        x = self.embed(x)
        x = self.linear(x)
        return x

model = TinyGPT(vocab_size)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
loss_fn = nn.CrossEntropyLoss()

losses = []

for step in range(300):
    idx = torch.randint(len(data)-1, (16,))
    x = torch.stack([data[i:i+1] for i in idx])
    y = torch.stack([data[i+1:i+2] for i in idx])

    logits = model(x)
    loss = loss_fn(logits.view(-1, vocab_size), y.view(-1))

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    losses.append(loss.item())

    if step % 50 == 0:
        print("step", step, "loss", loss.item())
```

```python
plt.plot(losses)
plt.title("Training Loss Curve")
plt.xlabel("Step")
plt.ylabel("Loss")
plt.show()
```

```python
text = ""

for term,desc in data:
    text += term + " : " + desc + "\n"

print(text)
```

```python
chars = sorted(list(set(text)))

stoi = {ch:i for i,ch in enumerate(chars)}
itos = {i:ch for ch,i in stoi.items()}

def encode(s):
    return [stoi[c] for c in s]

def decode(l):
    return ''.join([itos[i] for i in l])

vocab_size = len(chars)

print(vocab_size)
```

```python
encoded = torch.tensor(encode(text), dtype=torch.long)

x = encoded[:-1]
y = encoded[1:]

x = x.unsqueeze(0)
y = y.unsqueeze(0)
```

```python
class CyberLLM(nn.Module):

    def __init__(self,vocab_size):
        super().__init__()

        self.embedding = nn.Embedding(vocab_size,64)
        self.rnn = nn.GRU(64,128,batch_first=True)
        self.fc = nn.Linear(128,vocab_size)

    def forward(self,x):

        x = self.embedding(x)
        out,_ = self.rnn(x)
        out = self.fc(out)

        return out
```

```python
model = CyberLLM(vocab_size)

optimizer = torch.optim.Adam(model.parameters(), lr=0.003)

loss_fn = nn.CrossEntropyLoss()
```

```python
losses = []

for epoch in range(300):

    output = model(x)

    loss = loss_fn(output.view(-1,vocab_size), y.view(-1))

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    losses.append(loss.item())

    if epoch % 50 == 0:
        print("epoch:",epoch,"loss:",loss.item())
```

```python
plt.plot(losses)
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```
